```python
# Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
query2 = "select *, case when dayname(rental_date) = 'Monday' then 'Workday' \
when dayname(rental_date) = 'Tuesday' then 'Workday' \
when dayname(rental_date) = 'Wednesday' then 'Workday' \
when dayname(rental_date) = 'Thursday' then 'Workday' \
when dayname(rental_date) = 'Friday' then 'Workday' \
when dayname(rental_date) = 'Saturday' then 'Weekend' \
when dayname(rental_date) = 'Sunday' then 'Weekend' \
else 'Unknown' end as 'Day_type', dayname(rental_date) as weekday \
from sakila.rental;"
data2 = pd.read_sql_query(query2, engine)
data2.head()     
```

```python
# Get all films with ARMAGEDDON in the title.
query3 = 'select title from sakila.film where title like "%%ARMAGEDDON%%";'
data3 = pd.read_sql_query(query3, engine)
data3
```

```python
# Get 10 the longest films.
query4 = 'select title, length from sakila.film \
order by length desc limit 10;'
data4 = pd.read_sql_query(query4, engine)
data4.head(10)
```

```python
# How many films include Behind the Scenes content?
query5 = '''select film_id, count(special_features) from sakila.film
where special_features like "%%behind the scenes%%" group by film_id;'''
data5 = pd.read_sql_query(query5, engine)
data5
```

```python
# Which kind of movies (rating) have a mean duration of more than two hours?
query6 = ''' select rating, avg(length) as avg_length_by_rating from sakila.film group by rating having avg(length)>120;'''    
data6 = pd.read_sql_query(query6, engine)
data6.head()
```
